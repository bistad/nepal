```python
Data = []
d2color = {}
for key,value in Nepal.items():
    PV = key
    dists = value
    for key,value in dists.items():
        district = key
        color = "%06x" % random.randint(0, 0xFFFFFF)
        d2color.update({district: "#"+color})
        #print(district,":","#"+color)
        MP = value
        for key,value in MP.items():
            mcp = key
            MCP = mcp.replace(" ","_")
            wards = value
            for w in wards:
                print(PV,district,MCP,w.replace(" ", "_"))
                Data.append({"id":random.choice([250,500,1000,2000,5000]) , "value":PV+"-"+district+"-"+MCP+"-"+str(w.replace(" ", "_"))})
```

```python
with open("d2color.json", 'w') as fw:
    json.dump(d2color,fw)
```
